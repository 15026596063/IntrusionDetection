```python
def dummy_df(df):
    todummylist=['protocol_type','service','flag']
    for x in todummylist:
        dummies=pd.get_dummies(df[x],dummy_na=False)
        df=df.drop(x,1)
        df=pd.concat([df,dummies],axis=1)
    return df
```

```python
df=dummy_df(df)
df.head(5)
```

```python
#Roc -> Receiver Operating Characterestic. Graphical plot that illustrates diagnostic ability of binary classifier as its discrimination 
#threshold is varied. 
#AUC - > Percent of area that is under the curve. The higher, the better. 
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
X = df.drop ('class', 1) # Dropping target, train_features = train.iloc[:,:41]
#X = X.drop ('num_outbound_cmds', 1)
#X = X.drop ('is_host_login', 1)
y = df['class'] #train_target = train.class 
X.shape

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
from sklearn.metrics import roc_curve
keras_model=create_model()
y_pred_keras=keras_model.predict(X_test).ravel()
fpr_keras,tpr_keras,thresholds_keras=roc_curve(y_test,y_pred_keras)
```

```python
#AUC Value:
from sklearn.metrics import auc
auc_keras=auc(fpr_keras,tpr_keras)
```

```python
from sklearn.ensemble import RandomForestClassifier #To compare RandomForestClassifier with Keras classifier in same plot.
rf=RandomForestClassifier(max_depth=3,n_estimators=10)
rf.fit(X_train,y_train)
y_pred_rf=rf.predict_proba(X_test)[:,1]
fpr_rf,tpr_rf,thresholds_rf=roc_curve(y_test,y_pred_rf)
auc_rf=auc(fpr_rf,tpr_rf)
```

```python
plt.figure(1)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_keras, tpr_keras, label='Keras (area = {:.3f})'.format(auc_keras))
plt.plot(fpr_rf, tpr_rf, label='RF (area = {:.3f})'.format(auc_rf))
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.legend(loc='best')
plt.show()
# Zoom in view of the upper left corner.
plt.figure(2)
plt.xlim(0, 0.2)
plt.ylim(0.8, 1)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_keras, tpr_keras, label='Keras (area = {:.3f})'.format(auc_keras))
plt.plot(fpr_rf, tpr_rf, label='RF (area = {:.3f})'.format(auc_rf))
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve (zoomed in at top left)')
plt.legend(loc='best')
plt.show()
```
